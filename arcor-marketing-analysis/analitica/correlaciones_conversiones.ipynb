# Correlaciones entre inversión y conversiones
Análisis de correlación y regresión simple entre gasto y conversiones.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white')
```

```python
ads = pd.read_csv('analitica/datos_publicidad.csv')
corr_df = ads[['impresiones','clicks','gasto_ars','conversiones','ingresos_ars']].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr_df, annot=True, cmap='vlag', fmt='.2f')
plt.title('Correlación entre métricas de publicidad (simulado)')
plt.tight_layout()
plt.show()
```

```python
# Regresión gasto vs conversiones
import statsmodels.api as sm
X = ads['gasto_ars']
y = ads['conversiones']
X2 = sm.add_constant(X)
modelo = sm.OLS(y, X2).fit()
print(modelo.summary())
```
